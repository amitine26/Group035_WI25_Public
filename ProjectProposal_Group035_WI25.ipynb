# Research Question

- What factors increased the rate of car crashes in Chicago in 2018?

- **Subquestions:**
   - How did environmental conditions (weather, lighting, and roadway surface) impact the rate of car crashes in Chicago during 2018?
   - How did the time of day, day of the week, and holiday periods influence the rate of car crashes in Chicago during 2018?
   - Did the geographic location within Chicago impact the likelihood of car crashes in 2018?


## Background and Prior Work


This study investigates the impact of environmental conditions, temporal patterns, and geographic distribution on the frequency of car crashes in Chicago during 2018. By analyzing these factors, our research aims to uncover patterns that could inform traffic management strategies and contribute to road safety improvements. 

**Temporal Patterns**: Research indicates that the timing of car accidents can vary significantly depending on the day of the week and time of day. A comprehensive study from Dropplr suggests that the highest frequency of accidents occurred during the afternoon rush hours of 2:00 PM to 5:00 PM, with Fridays being particularly prone to collisions. This pattern raises questions about the role of traffic volume and driver behavior at these times, which our study will explore further<a href="#cite_note-1">[1]</a>.

**Environmental Conditions**: Previous studies have shown that weather conditions play a significant role in road safety. For instance, the Kaggle Chicago Car Crashes Data Analysis found that overcast and partially cloudy weather during spring and early summer resulted in the highest rate of car crashes with fixed objects, pedestrians, and angle being the categories with more severe accidents<a href="#cite_note-2">[2]</a>.

**Geographic Distribution**: Our research will also examine how the geographic location within Chicago impacts the likelihood of car crashes. We hypothesize that different areas within the city may exhibit varied crash rates due to factors such as road design, traffic volume, and local enforcement of traffic laws. Understanding the spatial distribution of accidents can help identify high-risk areas, guiding targeted interventions and safety improvements. By mapping accident data across Chicago, we aim to pinpoint where the highest concentrations of accidents occur and analyze potential underlying causes related to the geographic characteristics of these areas.

**Chicago’s Traffic Safety Efforts**: The Kaggle Chicago Car Crashes Data Analysis found that the key factors of causing road accidents are: issues of yielding right of way, tailgating, overtaking, not adjusting speed, and improper backing maneuvers. The city’s ongoing traffic safety initiatives, including the Vision Zero plan, aim to reduce traffic fatalities by addressing factors like vehicle speed and reckless driving. Our study will examine how these efforts have possibly influenced accident rates and the persistence of high-risk zones throughout the city<a href="#cite_note-2">[3]</a>.

# Hypothesis


- We hypothesize that in Chicago during 2018, the rate of car crashes was significantly higher during periods of snowfall and in areas with inadequate lighting. Due to the compounded adverse effects of reduced visibility and traction, we predict that more crashes would occur during these specific conditions.

- We hypothesize that in Chicago during 2018, the rate of car crashes was significantly higher during the evening hours, on weekends and holidays. This increase can be attributed to higher road congestion as people are more likely to be going out, alongside an increase in impaired driving due to alcohol consumption during these times.

- We hypothesize that in Chicago during 2018, the likelihood of car crashes was significantly higher near the on and off ramps of highways and freeways, where frequent merging and diverging traffic increases the potential for accidents. Additionally, areas with higher traffic volumes and intersections with complex traffic patterns may also exhibit a higher incidence of crashes.


# Data

1. An ideal dataset to answer our research question would have at least 100,000 data points of car crashes from the city of Chicago during the year of 2018. This data should be collected by the city of Chicago or one of the departments within Chicago that keeps track of car crashes. Ideally, the dataset would contain information on the weather condition, lighting condition, and road conditions during the time of the car crash. Additionally, we would want to be able to know the exact date and time that the crash was reported as that would allow us to determine if there is an increase in car crashes during holidays. It would be nice to know the latitude and longitude coordinates of the location of the crash to determine if there are specific locations where crashes are more likely to have occurred. Lastly, it would be a good cross reference if the dataset contains information on blood alcohol content (BAC) of the driver(s) involved in the crash, as then we would be able to see if a DUI was a potential cause for the crash.
