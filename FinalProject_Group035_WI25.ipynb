# Abstract

Car crashes are a major public safety concern, and understanding the factors contributing to their occurrence can help improve traffic management and road safety. This study investigates the impact of environmental conditions, temporal patterns, and geographic distribution on car crashes in Chicago from 2016 to 2024. Using a dataset of over 900,000 reported crashes, we conducted exploratory data analysis (EDA) to identify trends related to weather conditions, lighting, roadway surface conditions, time of day, and geographic location. 
 
 Contrary to our initial hypothesis, we found that the majority of crashes occurred under clear weather and daylight conditions rather than during adverse weather. Additionally, crashes peaked during morning and evening rush hours, with Fridays showing the highest crash rates. Geographic analysis revealed persistent crash hotspots, particularly in central Chicago, where traffic density and complex intersections may contribute to higher accident frequencies. 
 
 These findings suggest that traffic volume and driver behavior play a more significant role in crash frequency than environmental conditions alone. The study highlights the need for targeted safety interventions, such as improved traffic control measures in high-risk areas and strategic enforcement during peak traffic times. Future research could further explore the relationship between human factors, infrastructure, and crash severity to enhance urban road safety strategies. 
 
 Our findings revealed several unexpected insights. While we initially hypothesized that crashes would be more frequent in adverse weather conditions (rain, snow, and low lighting), the majority of crashes actually occurred during clear weather and daylight. This suggests that higher traffic volume during these conditions plays a more significant role in crash frequency than poor visibility or road traction. Additionally, crashes peaked during morning and evening rush hours, particularly on Fridays. Geographically, we identified consistent crash hotspots, particularly around central Chicago.

```python
print(chicago_crash['WEATHER_CONDITION'].value_counts())
```

```python
# Subsetting car crashes from to include weather condition and crash date date.
chicago_crash_weather = chicago_crash.dropna(subset=['WEATHER_CONDITION', 'CRASH_DATE'])

# Group by WEATHER_CONDITION and calculate the number of crashes and crash rate
weather_crash_counts = chicago_crash_weather.groupby('WEATHER_CONDITION').size().reset_index(name='Crash_Count')

# Calculate the total number of crashes for rate calculations
total_crashes = chicago_crash_weather.shape[0]

# Calculate crash rate per weather condition (Crash Count / Total Crashes)
weather_crash_counts['Crash_Rate'] = weather_crash_counts['Crash_Count'] / total_crashes

# Sort by crash rate
weather_crash_counts = weather_crash_counts.sort_values(by='Crash_Rate', ascending=False)

# Print out the crash count and rate by weather condition
print(weather_crash_counts)

# Visualization
plt.figure(figsize=(10, 6))
sns.barplot(x='Crash_Rate', y='WEATHER_CONDITION', data=weather_crash_counts, hue='WEATHER_CONDITION', palette='viridis', legend=False)
plt.title('Crash Rate by Weather Condition')
plt.xlabel('Crash Rate')
plt.ylabel('Weather Condition')
plt.show()
```

```python
# Clean data by removing rows with missing values for environmental factors
chicago_crash_environment = chicago_crash.dropna(subset=['WEATHER_CONDITION', 'LIGHTING_CONDITION', 'ROADWAY_SURFACE_COND', 'ROAD_DEFECT'])

# Function to group rare categories
def collapse_categories(column, min_count=500):  
    category_counts = chicago_crash_environment[column].value_counts()
    common_categories = category_counts[category_counts >= min_count].index
    chicago_crash_environment[column] = chicago_crash_environment[column].apply(lambda x: x if x in common_categories else 'Other')

# Apply category grouping
for col in ['WEATHER_CONDITION', 'LIGHTING_CONDITION', 'ROADWAY_SURFACE_COND', 'ROAD_DEFECT']:
    collapse_categories(col)

# Create contingency table for environmental factors
contingency_table = chicago_crash_environment.groupby(
    ['WEATHER_CONDITION', 'LIGHTING_CONDITION', 'ROADWAY_SURFACE_COND', 'ROAD_DEFECT']
).size().unstack(fill_value=0)

# Perform Chi-Square Test with Monte Carlo simulation
chi2, p, dof, expected = stats.chi2_contingency(contingency_table, lambda_='log-likelihood')

# Print Chi-Square results
print(f'Chi-Square Statistic: {chi2}')
print(f'P-value: {p}')
print(f'Degrees of Freedom: {dof}')

# Define Cramér's V function
def cramers_v(chi2, n, contingency_table):
    k = min(contingency_table.shape)  # The smaller of the number of rows or columns
    return np.sqrt(chi2 / (n * (k - 1)))

# Create contingency table for environmental factors (using previously cleaned data)
contingency_table_env = pd.crosstab(
    chicago_crash_environment['WEATHER_CONDITION'], 
    [chicago_crash_environment['LIGHTING_CONDITION'], 
     chicago_crash_environment['ROADWAY_SURFACE_COND'], 
     chicago_crash_environment['ROAD_DEFECT']]
)

# Perform Chi-Square Test for Environmental Factors
chi2_env, p_env, dof_env, expected_env = stats.chi2_contingency(contingency_table_env)

# Calculate Cramér's V for Environmental Factors
n_env = contingency_table_env.sum().sum()  # Total number of crashes for environmental factors
cramers_v_env = cramers_v(chi2_env, n_env, contingency_table_env)

# Print the result
print()
print(f"Cramér's V for Environmental Factors: {cramers_v_env}")
```

This analysis examined the relationship between environmental conditions (weather, lighting, roadway surface, and road defects) and crash occurrence using a Chi-Square test. The results showed a highly significant association (χ² = 199,981.65, p = 0.0) with 2,172 degrees of freedom, indicating a strong dependency between these factors. The effect size, measured by Cramér’s V (V = 0.445), suggests a moderate to strong relationship. This highlights the importance of environmental factors in crash risk and the need for targeted safety measures under adverse conditions.

```python
# Aggregate the data by day of the week and hour to get the count of crashes
crash_long_data = chicago_crash.groupby(['CRASH_DAY_OF_WEEK', 'CRASH_HOUR']).size().reset_index(name='count')

# Fit an ANOVA model (including interaction term)
model = ols('count ~ CRASH_DAY_OF_WEEK + CRASH_HOUR + CRASH_DAY_OF_WEEK:CRASH_HOUR', data=crash_long_data).fit()

# Perform the ANOVA test
anova_results = anova_lm(model)

# Display the ANOVA results
print(anova_results)

# Calculate Eta-squared for each factor (CRASH_DAY_OF_WEEK, CRASH_HOUR, CRASH_DAY_OF_WEEK:CRASH_HOUR)
# Eta-squared = sum_sq_factor / (sum_sq_factor + sum_sq_residual)

# Eta-squared for CRASH_DAY_OF_WEEK
eta_squared_day_of_week = anova_results['sum_sq']['CRASH_DAY_OF_WEEK'] / (anova_results['sum_sq']['CRASH_DAY_OF_WEEK'] + anova_results['sum_sq']['Residual'])
f_squared_day_of_week = eta_squared_day_of_week / (1 - eta_squared_day_of_week)  # Cohen's f²

# Eta-squared for CRASH_HOUR
eta_squared_hour = anova_results['sum_sq']['CRASH_HOUR'] / (anova_results['sum_sq']['CRASH_HOUR'] + anova_results['sum_sq']['Residual'])
f_squared_hour = eta_squared_hour / (1 - eta_squared_hour)  # Cohen's f²

# Eta-squared for the interaction term CRASH_DAY_OF_WEEK:CRASH_HOUR
eta_squared_interaction = anova_results['sum_sq']['CRASH_DAY_OF_WEEK:CRASH_HOUR'] / (anova_results['sum_sq']['CRASH_DAY_OF_WEEK:CRASH_HOUR'] + anova_results['sum_sq']['Residual'])
f_squared_interaction = eta_squared_interaction / (1 - eta_squared_interaction)  # Cohen's f²

# Print results
print()
print(f'Cohen\'s f-squared for CRASH_DAY_OF_WEEK: {f_squared_day_of_week}')
print()
print(f'Cohen\'s f-squared for CRASH_HOUR: {f_squared_hour}')
print()
print(f'Cohen\'s f-squared for CRASH_DAY_OF_WEEK:CRASH_HOUR interaction: {f_squared_interaction}')
```

To analyze how crash frequency varies by the day of the week and hour of the day, an ANOVA test was performed. The results showed that crash day of the week had no significant effect (p = 0.945), while crash hour was highly significant (p < 1.17 × 10⁻⁹), indicating that crash likelihood strongly depends on the time of day. The interaction effect between day and hour was also non-significant (p = 0.736), suggesting a consistent daily crash pattern. The effect size for crash hour (f² = 0.254) was moderate, emphasizing the need for time-based safety interventions.

```python
# Crash Count on Holidays vs. Non-Holidays

# Define Fixed Holidays
fixed_holidays = ["01-01", "02-14", "03-17", "07-04", "10-31", "12-24", "12-25", "12-31"]

# Define Movable Holidays
def get_movable_holidays(year):
    """
    Returns a list of dynamically computed holidays for a given year.
    """
    # Mardi Gras (47 days before Easter Sunday)
    mardi_gras = (easter(year) - timedelta(days=47)).strftime("%m-%d")

    # Easter Sunday
    easter_sunday = easter(year).strftime("%m-%d")

    # Memorial Day (Last Monday in May)
    last_monday_may = max(
        [datetime(year, 5, d) for d in range(25, 32) if datetime(year, 5, d).weekday() == 0]
    ).strftime("%m-%d")

    # Labor Day (First Monday in September)
    first_monday_sept = min(
        [datetime(year, 9, d) for d in range(1, 8) if datetime(year, 9, d).weekday() == 0]
    ).strftime("%m-%d")

    # Thanksgiving (Fourth Thursday in November)
    fourth_thursday_nov = min(
        [datetime(year, 11, d) for d in range(22, 30) if datetime(year, 11, d).weekday() == 3]
    ).strftime("%m-%d")

    # Black Friday (Day after Thanksgiving)
    black_friday = (datetime(year, 11, int(fourth_thursday_nov.split('-')[1])) + timedelta(days=1)).strftime("%m-%d")

    return [mardi_gras, easter_sunday, last_monday_may, first_monday_sept, fourth_thursday_nov, black_friday]

# Ensure CRASH_DATE_ONLY is a Datetime Object

# Compute Holidays for Each Year in Dataset
holiday_dict = {year: fixed_holidays + get_movable_holidays(year) for year in chicago_crash['CRASH_YEAR'].unique()}

# Check if Each Crash Date is a Holiday
chicago_crash['IS_HOLIDAY'] = chicago_crash.apply(
    lambda row: row['CRASH_DATE_ONLY'].strftime("%m-%d") in holiday_dict[row['CRASH_YEAR']], axis=1
)

# Plot
holiday_counts = chicago_crash['IS_HOLIDAY'].value_counts()
plt.figure(figsize=(6, 4))
sns.barplot(x=holiday_counts.index, y=holiday_counts.values, hue=holiday_counts.index, palette='viridis', legend=False)
plt.xticks([0, 1], ['Non-Holiday', 'Holiday'])
plt.title('Crash Count on Holidays vs. Non-Holidays', fontsize=14)
plt.xlabel('Holiday Status', fontsize=12)
plt.ylabel('Number of Crashes', fontsize=12)
plt.grid(axis='y')
plt.show()
```

This graph shows the number of crashes that happened on a holiday vs a regular day. We predicted that more crashes would happen on holidays however, this graph shows us that an overwhelming amount happen on regular days with over 800,000 crashes whereas on holidays the crash count was under 100,000. Although we believed holidays would lead to more crashes due to higher alcohol consumption and people just generally in a different more distracted mind set, this result also makes sense. Holidays are just a few days during the year, and although it’s possible to have a higher crash rate on these days, that doesn’t mean that the total crash number will be higher than the number of crashes that happen on all the rest of the days of the year.

```python
# Separate crashes into holiday and non-holiday groups
holiday_crashes = chicago_crash[chicago_crash['IS_HOLIDAY'] == 1].groupby('CRASH_DATE_ONLY').size()
non_holiday_crashes = chicago_crash[chicago_crash['IS_HOLIDAY'] == 0].groupby('CRASH_DATE_ONLY').size()

# Perform Welch’s t-test
t_stat, p_value = ttest_ind(holiday_crashes, non_holiday_crashes, equal_var=False)

# Convert p-value to high precision
p_value_decimal = Decimal(p_value)

print(f't-Statistic: {t_stat:.4f}')
print(f'p-value: {p_value_decimal:.10E}')

# Cohen's d for Welch’s t-test
mean_diff = np.mean(holiday_crashes) - np.mean(non_holiday_crashes)  # Difference in means
pooled_std = np.sqrt(((np.std(holiday_crashes)**2) + (np.std(non_holiday_crashes)**2)) / 2)  # Pooled standard deviation
cohens_d = mean_diff / pooled_std  # Cohen's d

# Print results
print()
print(f'Cohen\'s d: {cohens_d}')
```

A Welch’s t-test was conducted to compare crash rates on holidays versus non-holidays, revealing significantly fewer crashes on holidays (t = -4.78, p ≈ 4.51 × 10⁻⁶). The effect size, measured by Cohen’s d (d = -0.44), indicates a small to moderate decrease in crash frequency on holidays. This suggests that lower traffic volumes or altered driving behaviors during holidays contribute to reduced crashes. Understanding these patterns could help optimize traffic control measures during peak holiday periods.

### Geographic Location

In this section, we analyze the impact of geographic location on car crashes in Chicago from 2016 to 2024. We first clean and process the dataset by extracting latitude and longitude data and identifying high-crash locations (hotspots). Using density heatmaps we visualize crash concentration across different areas.

```python
# Define years from the dataset
years = sorted(chicago_crash['CRASH_YEAR'].unique())

# Define the geographic crash locations
geo_crash = chicago_crash[['CRASH_DATE', 'LATITUDE', 'LONGITUDE', 'CRASH_YEAR']].dropna()
geo_crash = geo_crash[
    (geo_crash['LATITUDE'].between(41.6, 42.1)) &
    (geo_crash['LONGITUDE'].between(-87.9, -87.5))
]

# Create yearly heatmaps
for year in years:
    plt.figure(figsize=(10, 6))
    # Get crashes for the given year (fixing the column name)
    geo_year = geo_crash[geo_crash['CRASH_YEAR'] == year]

    # Adjust sample size to be the minimum of 50,000 or the available data
    sample_size = min(50000, len(geo_year))

    # Sample data (only if there's enough)
    geo_sample = geo_year.sample(n=sample_size, random_state=42)

    # Generate heatmap
    sns.kdeplot(x=geo_sample['LONGITUDE'], y=geo_sample['LATITUDE'], cmap='Reds', fill=True, alpha=0.7)

    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    plt.title(f'Heatmap of Car Crash Density in Chicago ({year})')
    plt.show()
```

These graphs show the heat map of the car's geographic location every year from 2016 to 2024. From this, we can see that over all years, there’s one spot around 41.9 and -87.65 that has consistently been a major crash area. It looks to have minorly changed shape and size over the years but is overall pretty consistent. One surprising thing from these graphs is that from 2020 to 2023, there looks to be an increase in car crashes throughout Chicago as all areas on the map got darker, especially at 41.9 and -87.75. This was right at the start of the pandemic, so maybe some factors from that led to this large increase in another area. However, after 2020, the second area slowly started to decrease in size until about 2023, when it was almost the same size it was in 2016. An observation is the consistency of crash hotspots over the years. The most significant hotspot appears around latitude 41.9 and longitude -87.65, which likely corresponds to downtown Chicago or a major traffic corridor. Another secondary hotspot emerges intermittently around latitude 41.75 to 41.8, suggesting another high-risk location, possibly near major highways or intersections. These areas likely experience a high volume of traffic, frequent stop-and-go movement, and complex intersections, all of which contribute to the high crash rates.

```python
# Create a Table for the Top 10 Crash Hotspots in Chicago
geo_crash['LAT_LONG'] = geo_crash['LATITUDE'].round(3).astype(str) + ', ' + geo_crash['LONGITUDE'].round(3).astype(str)

# Count crashes per location (hotspots)
hotspots = geo_crash['LAT_LONG'].value_counts().reset_index()
hotspots.columns = ['LAT_LONG', 'CRASH_COUNT']

# Display the top 10 crash hotspots
print('Top 10 Crash Hotspots in Chicago:')
print(hotspots.head(10))
```

This table shows the top ten crash spots in Chicago. We can see that most of these spots are somewhere around 41.7 and -87.6. Although the ten points vary a bit between these coordinates, all are within the one major spot we saw when looking at the breakdown of years through heatmaps.

```python
# Round lat/long to group crashes into geographic bins
geo_crash['LAT_LONG'] = geo_crash['LATITUDE'].round(2).astype(str) + ', ' + geo_crash['LONGITUDE'].round(2).astype(str)

# Create a contingency table of crash counts per location per year
contingency_table = pd.crosstab(geo_crash['CRASH_YEAR'], geo_crash['LAT_LONG'])

# Perform Chi-Square Test
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)

# Print results
print(f'Chi-Square Statistic: {chi2}')
print(f'P-value: {p}')
print(f'Degrees of Freedom: {dof}')

# Function to compute Cramér's V
def cramers_v(chi2, n, contingency_table):
    k = min(contingency_table.shape)  # The smaller of the number of rows or columns
    return np.sqrt(chi2 / (n * (k - 1)))

# Calculate Cramér's V for Location-Year
n_location_year = contingency_table.sum().sum()  # Total number of crashes
cramers_v_location_year = cramers_v(chi2, n_location_year, contingency_table)

# Print the result
print()  
print(f"Cramér's V for Location-Year: {cramers_v_location_year}")
```

To determine whether crash locations change significantly over time, a Chi-Square test was conducted on crash counts per geographic bin across different years. The results showed a significant association (χ² = 22,612.83, p = 0.0) with 5,672 degrees of freedom, but the effect size (V = 0.056) was very small. This suggests that while crash locations do vary slightly over time, the overall distribution of high-risk areas remains relatively stable. These findings reinforce the need for long-term, location-based safety improvements.
